# Exploratory Data Analysis (EDA) untuk Prediksi Potabilitas Air

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

**Import library yang dibutuhkan** seperti `pandas` untuk manipulasi data, `seaborn` dan `matplotlib` untuk visualisasi, serta `os` untuk manajemen file.

```
# Load dataset
df = pd.read_csv('model/water_potability.csv')
df.fillna(df.mean(), inplace=True)
```

**Membaca dataset** dari file CSV dan **mengisi nilai kosong dengan rata-rata** kolom masing-masing agar data siap dianalisis.

```
os.makedirs('static/img', exist_ok=True)
```

**Membuat folder untuk menyimpan hasil visualisasi**, jika belum tersedia.

```
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Antar Fitur')
plt.tight_layout()
plt.savefig('static/img/heatmap.png')
plt.close()
```

**Visualisasi korelasi antar fitur** menggunakan heatmap untuk melihat hubungan antar variabel numerik.

```
plt.figure(figsize=(6, 4))
sns.countplot(x='Potability', data=df, palette='pastel')
plt.title('Distribusi Potability (Layak vs Tidak Layak)')
plt.xlabel('Potability')
plt.ylabel('Jumlah')
plt.savefig('static/img/potability_dist.png')
plt.close()
```

**Distribusi data target `Potability`** untuk melihat proporsi air yang layak dan tidak layak konsumsi.

```
for col in df.columns[:-1]:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribusi {col}')
    plt.xlabel(col)
    plt.ylabel('Frekuensi')
    plt.tight_layout()
    plt.savefig(f'static/img/hist_{col}.png')
    plt.close()
```

**Distribusi masing-masing fitur** ditampilkan dalam bentuk histogram untuk memahami karakteristik tiap variabel.
